```python
# Device used for normal execution
device_name=tf.config.experimental.list_logical_devices('GPU')
print(" Using Device : ", 'GPU' if device_name else 'CPU')
```

```python
#This function generates the embeddings
def get_embeddings_in_batches(texts,device_name1,batch_size=16):
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size] #Use of batch processing
        inputs = tokenizer(batch_texts, return_tensors="tf", padding=True, truncation=True, max_length=512)
        with tf.device('/GPU:0' if device_name1 else '/CPU:0'):
            outputs = embedding_model(inputs).last_hidden_state
            batch_embeddings = tf.reduce_mean(outputs, axis=1)
        all_embeddings.extend(batch_embeddings.numpy())
    return np.array(all_embeddings)
# Customized token based splitter for token chunking method
def Tokentextsplitter(para,chunk_size=100,overlap_percentage=0.5):
  overlap_size=int(overlap_percentage*100)
  # Dividing paragraph into necessary tokens
  tokens = re.findall(r"\w+|[^\s\[\]]", para)
  arr=[]
  j=0
  t=[]
  while(j<len(tokens)):
        t.append(' '.join(tokens[j:j+chunk_size])) # Joining and appending tokens based on chunk size. For example chunk size = no of tokens added in
        j=j+chunk_size-overlap_size # To ensure overlapping charecters in all sentences
  return t
```

```python
# Preprocessing along with recursive based chunking
def PreprocessingRChunking(passage,file_path,csize=200,overlap_percentage=0.50):
  #Setting size of overlap
  coverlap=int(overlap_percentage*csize)
  print(f'Loaded {len(passage)} abstracts of MS-Macro QA')
  cleaned_texts=[]
  temp=[]
  all_chunks=[]
  #Accesing each paragraph of content
  for context in passage:
    # Cleaning data
    text=simple_preprocess(str(context), deacc=True)
    cleaned_text=' '.join(text)
    #Calling Recursive Chunking module from Langchain
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=csize,chunk_overlap=coverlap)
    #Splitting document
    doc=text_splitter.create_documents([cleaned_text])
    for i in doc:
      temp.append(i.page_content)
    cleaned_texts.append(cleaned_text)
    #Chunked text will be stored here together
    all_chunks.extend(temp)
    temp=[]
  #Converting text to embeddings
  embeddings=get_embeddings_in_batches(all_chunks,device_name)
  print(f"Generated embeddings complete.")
  #If file path is mentioned then embeddings can be saved
  if(file_path!=''):
    pickle.dump(embeddings,open(file_path,'wb'))
  return embeddings,all_chunks
```

```python
# Preprocessing along with token based chunking
def PreprocessingTChunking(passage,file_path,csize=200,overlap_percentage=0.50):
  #Similar to previous function in terms of preprcessing
  print(f'Loaded {len(passage)} abstracts of MS-Macro QA')
  cleaned_texts=[]
  all_chunks=[]
  for context in passage:
    text=simple_preprocess(str(context), deacc=True)
    cleaned_text=' '.join(text)
    temp=Tokentextsplitter(cleaned_text,csize,overlap_percentage)
    cleaned_texts.append(cleaned_text)
    all_chunks.extend(temp)
    temp=[]
  embeddings=get_embeddings_in_batches(all_chunks,device_name)
  print(f"Generated embeddings complete.")
  if(file_path!=''):
    pickle.dump(embeddings,open(file_path,'wb'))
  return embeddings,all_chunks
```

```python
def opFAISS(embeddings,file_path1):
  embedding_dim = embeddings.shape[1]
  index = faiss.IndexFlatL2(embedding_dim)
  embedding_matrix = embeddings.reshape(-1, embedding_dim).astype('float32')
  index.add(embedding_matrix)
  if(file_path1!=''):
      pickle.dump(index,open(file_path1,'wb'))
  return index
```

```python
def retrieve_similar_abstracts(index,embeddings,cleaned_texts,query, top_k=3):
    query_embedding = get_embeddings_in_batches([query],device_name)
    distances, indices = index.search(query_embedding.reshape(1, -1), top_k)
    results = []
    for idx in indices[0]:
        results.append(cleaned_texts[idx])
    return results
```

```python
def generate_response(index,embe,cleaned_t,query,max_new_tokens=100):
    sample_query=query
    print('Sample Query:',sample_query)
    similar_abstracts = retrieve_similar_abstracts(index,embe,cleaned_t,sample_query)
    print(f' Retrive the Documnets  : {similar_abstracts}')
    input_text = f"User query: {query}\n\nContext:\n{similar_abstracts}\n\nAnswer:"
    inputs = generation_tokenizer(input_text, return_tensors="pt", padding=True, truncation=True).to("cuda")
    with torch.no_grad():
        outputs = generation_model.generate(inputs["input_ids"], max_new_tokens=max_new_tokens, num_return_sequences=1, use_cache=False)
    response_text = generation_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response_text
```

```python
def getting_anssc(inde,embedin,clean_t,ques):
  ans=[]
  for i in range(len(ques)):
    rs=generate_response(inde,embedin,clean_t,ques[i])
    Ans=rs.find("Answer")
    resp=rs[Ans+7:]
    ans.append(resp)
  return ans
```
